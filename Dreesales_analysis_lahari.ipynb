As you can see, there is a column in “Attribute Dataset” named as ‘Size’. This column contains the values in abbreviation format. Write a code in Python to convert the followings:

- M into  “Medium”
- L into  “Large”
- XL into “Extra large”
- free into “Free”
- S, s & small into “Small”.

Now once you are done with changes in the dataset, what is the value of the lowest percentage, the highest percentage and the percentage of Small size categories in the column named “Size”?



```python
import pandas as pd

# Assuming inp0 is your DataFrame
# Example DataFrame
inp0 = pd.DataFrame({
    'Size': ['M', 'L', 'XL', 'free', 'S', 's', 'small', 'S', 'M', 'L', 'XL']
})

# Mapping size abbreviations to their corresponding values
size_mapping = {
    'M': 'Medium',
    'L': 'Large',
    'XL': 'Extra large',
    'free': 'Free',
    'S': 'Small',
    's': 'Small',
    'small': 'Small'
}

# Applying the mapping to the 'Size' column
inp0['Size'] = inp0['Size'].map(size_mapping)

# Calculating percentages
total_rows = len(inp0)
percentage_small = (inp0['Size'].value_counts(normalize=True).get('Small', 0)) * 100
lowest_percentage = inp0['Size'].value_counts(normalize=True).min() * 100
highest_percentage = inp0['Size'].value_counts(normalize=True).max() * 100

# Print the results
print(f"Lowest Percentage: {lowest_percentage:.2f}%")
print(f"Highest Percentage: {highest_percentage:.2f}%")
print(f"Percentage of Small size categories: {percentage_small:.2f}%")
```
